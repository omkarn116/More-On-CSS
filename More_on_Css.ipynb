1. Create a navigation bar similar to the below-mentioned image.The navigation bar must contain 5 navigation links that are properly placed within the navigation bar using the CSS box model. The navigation items must change their text color on hover.

```python
/* HTML */
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Navigation Bar</title>
    <link rel="stylesheet" href="p.css">
</head>
<body>
    <nav class="navbar">
        <ul>
            <li><a href="#home">Home</a></li>
            <li><a href="#about">About Us</a></li>
            <li><a href="#contact">Contact</a></li>
            <li><a href="#blog">Blog</a></li>
            <li><a href="#login">Login</a></li>
        </ul>
    </nav>
</body>
</html>
```

```python
/* CSS */
body {
    margin: 0;
    font-family: Arial, sans-serif;
}

.navbar {
    background-color: #6495ED; /* A shade of blue similar to the image */
    padding: 15px 0; /* Padding top and bottom for the navbar itself */
    box-shadow: 0 2px 5px rgba(0,0,0,0.2); /* Optional: subtle shadow for depth */
}

.navbar ul {
    list-style: none; /* Removes bullet points from the list */
    margin: 0;
    padding: 0;
    display: flex; /* Uses Flexbox for easy alignment of links */
    justify-content: center; /* Centers the navigation items horizontally */
    align-items: center; /* Aligns items vertically in the middle */
    box-sizing: border-box; /* Ensures padding and border are included in the element's total width and height */
}

.navbar li {
    margin: 0 20px; /* Space between each navigation item */
    /* Using box model properties for spacing */
    /* You can also use padding on the <a> tag itself if you prefer more clickable area */
}

.navbar a {
    text-decoration: underline; /* Underline as shown in the image */
    color: #fff; /* White text color for the links */
    font-size: 18px; /* Adjust font size as needed */
    padding: 5px 0; /* Vertical padding for the clickable area of links */
    display: block; /* Makes the whole padding area clickable */
    transition: color 0.3s ease; /* Smooth transition for hover effect */
    box-sizing: border-box; /* Ensures padding and border are included */
}

.navbar a:hover {
    color: #f0f0f0; /* Lighter shade of white on hover */
}
```
